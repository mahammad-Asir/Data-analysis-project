```python
import pandas as pd
```

```python
data = {
    'Region': ['North', 'North', 'East', 'West', 'South', 'East', 'North', 'South', 'West', 'West', 'East', 'South'],
    'Category': ['Electronics', 'Clothing', 'Grocery', 'Electronics', 'Clothing', 'Grocery', 'Electronics', 'Grocery', 'Clothing', 'Electronics', 'Clothing', 'Electronics'],
    'Sales': [1200, 450, 300, 1100, 550, 250, 1300, 400, 600, 1050, 500, 1150],
    'Quantity': [5, 10, 15, 4, 12, 20, 6, 18, 9, 5, 11, 7]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
print(region_sales)
```

```python
category_performance = df.groupby('Category').agg({'Sales': 'mean', "Quantity": 'sum'})
print(category_performance)
```

```python
multi_analysis = df.groupby(['Region', 'Category'])['Sales'].agg(['sum', 'mean'])
print(multi_analysis)
```

```python
pivot_table = pd.pivot_table(
    df,
    values='Sales',
    index='Region',
    columns='Category',
    aggfunc='sum',
)
print(pivot_table)
```
